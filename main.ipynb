Diabetics prediction using logistic regression

```python
import matplotlib.pyplot as pyplot
import pandas as pd
import numpy as np
```

First we will implement the model manually without using Scikit Learn

```python
df = pd.read_csv("diabetes2.csv", sep=",")
df.head
```

```python
X = df.drop("Outcome",axis=1,inplace=False)
Y = df['Outcome'].values.reshape(-1, 1)  
```

Now we will standardize the Features for better prediction

```python
X = (X - X.mean()) / X.std()
X = np.insert(X, 0, 1, axis=1)  
```

Sigmoid (Logistic) Function

```python
def sigmoid(z):
    return (1 / (1 + np.exp(-z)))
```

```python
epochs = 1000
alpha = 0.01
m,n = X.shape
theta = np.zeros((n,1))
cost_history=[]
for i in range(epochs):
    h =  X.dot(theta)
    h = sigmoid(h)
    gradient = (1/m) * X.T.dot(h - Y)
    theta = theta - alpha * gradient
    
    
    
```

```python
def predict(X, theta):
    return (sigmoid(X.dot(theta)) >= 0.5).astype(int)
```

```python
y_pred = predict(X, theta)
accuracy = np.mean(y_pred == Y) * 100
print(f"Training Accuracy: {accuracy:.2f}%")
```

Now lets use scikit learn

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python

X = df.drop('Outcome', axis=1)
y = df['Outcome']

X = (X - X.mean()) / X.std()

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

Validating the model

```python

y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```
